```python
import pandas as pd
import numpy as np
from IPython.display import Image
import matplotlib.pyplot as plt
import networkx as nx
from networkx.algorithms import community
# Node2Vec
```

# Betweenness Centrality

## Normal

```python
# Load the adjacency matrix into a numpy array.
normal_mat = np.loadtxt('../Normal.txt', dtype=int)
for i in range(0,70):
        normal_mat[i][i] = 0
G_Normal = nx.from_numpy_matrix(normal_mat)
print("Number Of Edges: ",len(G_Normal.edges()))
print("Number Of Nodes: ",len(G_Normal.nodes()))
```

```python
D_Normal=nx.betweenness_centrality(G_Normal) 
```

```python
plt.figure(figsize=(20,5))
plt.bar(range(len(D_Normal)), list(D_Normal.values()), align='center')
plt.xticks(range(len(D_Normal)), list(D_Normal.keys()))
plt.show()
```
